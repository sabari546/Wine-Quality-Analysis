```python
# Identify numerical columns to apply IQR-based outlier removal
numerical_cols = cleaned_wine.select_dtypes(include=['float64', 'int64']).columns
numerical_cols
```

```python
# Remove outliers using IQR
Q1 = cleaned_wine[numerical_cols].quantile(0.25)  # 25th percentile (Q1)
Q3 = cleaned_wine[numerical_cols].quantile(0.75)  # 75th percentile (Q3)
IQR = Q3 - Q1  # Interquartile range (IQR)
```

```python
# Define bounds to identify outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
```

```python
cleaned_wine_no_outliers = cleaned_wine[~((cleaned_wine[numerical_cols] < lower_bound) | 
                                           (cleaned_wine[numerical_cols] > upper_bound)).any(axis=1)]
# without this '~' outliers only, not the cleaned data.
```

# Find regression of PH on citric acid

```python
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm


#extract data from dataset and assign it to x  and y
x = cleaned_wine_no_outliers["citric acid"]
y = cleaned_wine_no_outliers["pH"]

#adds constant column which helps to make prediction
X_const = sm.add_constant(x)

#create model and fit
model = sm.OLS(y, X_const).fit()

print(model.summary())

prediction = model.predict(X_const)
print("*************************************************************************")
print("predicted pH values:  ")
print(prediction.values)
```

# Is a certain type of wine (red or white) associated with higher quality?

```python
from scipy import stats

# Step 1: Separate the data into red and white wines
red_wine = cleaned_wine_encoded[cleaned_wine_encoded['color_white'] == False]  # Red wine (False in 'color_white')
white_wine = cleaned_wine_encoded[cleaned_wine_encoded['color_white'] == True]  # White wine (True in 'color_white')

# Step 2: Calculate the average quality for both red and white wines
avg_quality_red = red_wine['quality'].mean()
avg_quality_white = white_wine['quality'].mean()

# Step 3: Perform a t-test to compare the quality between red and white wines
t_stat, p_value = stats.ttest_ind(red_wine['quality'], white_wine['quality'], equal_var=False)

# Step 4: Display the results
print(f"Average quality of red wine: {avg_quality_red}")
print(f"Average quality of white wine: {avg_quality_white}")
print(f"T-statistic: {t_stat}")
print(f"P-value: {p_value}")

# Interpretation: If p-value < 0.05, we reject the null hypothesis that the means are equal.
```

# Is the confidence interval for the differences of the proportion of white wine with rating 3 and the proportion of red wine with rating 3.

```python
df2 = df1.drop_duplicates()
```

```python
df2['quality'].unique()
```

```python
df2
```

```python
import numpy as np

# Count the number of white and red wines with a rating of 3
count_white_q3 = df2[(df2['type'] == "white") & (df2['quality'] == 3)].shape[0]
count_red_q3 = df2[(df2['type'] == "red") & (df2['quality'] == 3)].shape[0]

# Total counts
total_white = df2[df2['type'] == 'white'].shape[0] #total row counts
total_red = df2[df2['type'] == 'red'].shape[0] #total row counts

# Calculate proportions
p1 = count_white_q3 / total_white
p2 = count_red_q3 / total_red

# Standard error of the difference in proportions
SE = np.sqrt((p1 * (1 - p1) / total_white) + (p2 * (1 - p2) / total_red)) 
```

```python
SE
```

```python
# Critical value for 95% confidence interval
z = 1.96  # for 95% confidence level

# Calculate the difference in proportions
difference = p1 - p2

# Calculate the confidence interval
CI_lower = difference - z * SE
CI_upper = difference + z * SE

print(f"Confidence Interval for the difference in proportions: ({CI_lower}, {CI_upper})")
```
